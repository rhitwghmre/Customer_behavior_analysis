NOW I WANT TO REMOVE THE NULL VALUES IN REVIEW RATING BUT IT CAN BE REMOVED BY MEAN,MEDIAN BUT WE CHOOSE MEDIAN BCOZ IT IS BETTER FOR OUTLIERS SO WE DO MEDIAN ON THE BASED OF CATEGORIES FOR DIFF CATEGORIES DIFF MEDIAN

```python
df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

we use .fillna(x.median()) bcoz fill null values with median

```python
#create a colummn age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'],q=4,labels=labels)
```

```python
df[['age','age_group']].head()
```

✔ qcut() divides the age column into 4 equal-sized groups (quartiles).
✔ Each group gets one of your labels:
Young Adult, Adult, Middle-aged, Senior
✔ Groups have equal number of people, not equal age range.

So the youngest 25% → Young Adult,
Next 25% → Adult,
Next 25% → Middle-aged,
Oldest 25% → Senior


```python
# create column purchase_frequency_days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

This code converts text frequencies (Weekly, Monthly, etc.) into numbers (days).

✔ Step 1: A dictionary (frequency_mapping) matches each word to a number

Weekly → 7

Monthly → 30

Fortnightly / Bi-Weekly → 14

Quarterly → 90

Annually → 365

✔ Step 2: .map() replaces the text in the column

It looks at frequency_of_purchases, replaces each word using the dictionary,
and creates a new column purchase_frequency_days.

✔ Final Output

Text frequencies become numeric days:

Weekly → 7

Monthly → 30

Fortnightly → 14

Quarterly → 90

Annually → 365




.map() uses a dictionary to convert text labels into numbers.
